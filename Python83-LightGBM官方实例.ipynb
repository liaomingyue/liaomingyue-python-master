## 导入包

```python
import lightgbm as lgb
import pandas as pd
from sklearn.metrics import mean_squared_error,roc_auc_score
```

## 加载数据

```python
print("Loading data...")
df_train=pd.read_csv('files/data/python83/regression.train.txt',header=None,sep='\t')
df_test=pd.read_csv('files/data/python83/regression.test.txt',header=None,sep='\t')
print(df_train.shape)
print(df_test.shape)
df_train.head()
```

数据格式为：第一列为1,0标签，作为y；其余列为特征，为X

```python
y_train=df_train[0].values
y_test=df_test[0].values
X_train=df_train.drop(0,axis=1).values
X_test=df_test.drop(0,axis=1).values
```

## 创建lightgbm格式的数据

```python
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
```

## 配置参数
参数以字典的形式

```python
params={
    'task':'train',
    'boosting_type':'gbdt',
    'objective':'binary',
    'metric':{'l2','auc'},
    'num_leaves':31,
    'learing_rate':0.05,
    'feature_fraction':0.9,
    'bagging_fraction':0.8,
    'bagging_freq':5,
    'verbose':0
}
```

```python
## 训练
```

```python
print('Start training...')
gbm=lgb.train(params,
             lgb_train,
             num_boost_round=20,
             valid_sets=lgb_eval,
             early_stopping_rounds=5)
```

## 保存模型

```python
print('Save model...')
gbm.save_model('files/data/python83/model.pkl')
```

## 预测

```python
print('Start predicting...')
y_pred=gbm.predict(X_test,num_iteration=gbm.best_iteration)
y_pred
```

## 评价

```python
# eval
print('The rmse of prediction is:', roc_auc_score(y_test, y_pred) ** 0.5)
```
