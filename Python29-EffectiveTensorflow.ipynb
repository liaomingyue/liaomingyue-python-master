# Tensorflow基础

Tensorflow与其他数值计算库最大的区别就是操作符号化。这个强大的概念使得Tensorflow可以处理各种事情（例如自动区分），而其他库无法实现这些功能（例如numpy）

我们以两个矩阵相乘为例，首先我们看看numpy如何实现

```python
import numpy as np

x=np.random.normal(size=[10,10])
y=np.random.normal(size=[10,10])
z=np.dot(x,y)
print(z)
```

接着我们使用Tensorflow实现同样的计算

```python
import tensorflow as tf

x=tf.random_normal([10,10])
y=tf.random_normal([10,10])
z=tf.matmul(x,y)
sess=tf.Session()

z_val=sess.run(z)

print(z_val)
```

不像numpy那样立即进行相乘并产生结果，Tensorflow仅给我们一个节点（在图中）的句柄，如果我们直接打印“z”,会出现下面的结果

```python
print(z)
```

因为两个输入(x和y)有了明确定义的形状（shape）,tensorflow能够推断出张量的形状以及它的类型，为了计算出张量的值，我们需要创建一个会话（session），然后使用Session.run()去评估
