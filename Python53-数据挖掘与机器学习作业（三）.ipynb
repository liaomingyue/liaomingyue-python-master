采样问题指的是给定一个特定的概率分布P(z)，得到一批符合这个概率分布的样本点。

采样的方法有很多，MCMC是其中的一类方法，意思是利用Mento Carlo蒙特卡罗方法和Markov Chain马尔科夫链完成采样。

```python
import matplotlib.pyplot as plt
import time
import  math
def random(num = 2000):
    M = 1000
    A = 3^4
    B = 1
    x_0 = time.time()
    result = [x_0]
    while len(result) < num:
      result.append((result[-1]*A+B) % M)
    result.pop(0)
    return result

result = random(2000)
xaixs = [i for i,j in enumerate(result)]
print(result[:10])
f = plt.figure(1)
p1 = plt.scatter(xaixs,result)
#p2 = plt.hist(result)
plt.show()
```
