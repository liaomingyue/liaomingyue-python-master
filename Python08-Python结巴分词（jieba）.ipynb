```python
import os
#读取指定路径下的所有文件，返回所文件接起来的内容
def read_content(content_path):
    #初始化内容为空
    content=''
    #print(os.listdir(file_path))
    files=os.listdir(content_path)
    for file in  :
        #拼接完整路径
        full_path=os.path.join(content_path,file)
        #判断是否为文件
        if os.path.isfile(full_path):
            if full_path.endswith('.txt'):
                print('正在加载歌词文本{}'.format(full_path))
                content+=open(full_path,'r',encoding='utf8').read()
                # 每首歌词之间用换行符分隔
                content += '\n'
            else:
                print('{}不是歌词文本\n'.format(full_path))
    print('加载歌词完毕\n')
    return content
content=read_content('C:/Users/Perfect/Desktop/lyric/')
print(content)
```

```python
import jieba.analyse


#使用jieba的textrank提取出1000个关键词及其比重
result=jieba.analyse.textrank(content,topK=1000,withWeight=True)
# print(result)
keywords = dict()
for i in result:
    keywords[i[0]] = i[1]
#     print(i[0])
print(keywords)
```

```python
from PIL import Image, ImageSequence
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator

# 初始化图片
image = Image.open('C:/Users/Perfect/Desktop/lyric/xiong.png')
graph = np.array(image)

# 生成云图，这里需要注意的是WordCloud默认不支持中文，所以这里需要加载中文黑体字库
wc = WordCloud(font_path='‪C:/Windows/Fonts/STFANGSO.TTF',
    background_color='white', max_words=100, mask=graph)
wc.generate_from_frequencies(keywords)
image_color = ImageColorGenerator(graph)
```

```python
# 显示图片
plt.imshow(wc)
plt.imshow(wc.recolor(color_func=image_color))
plt.axis("off") # 关闭图像坐标系
plt.show()
```
